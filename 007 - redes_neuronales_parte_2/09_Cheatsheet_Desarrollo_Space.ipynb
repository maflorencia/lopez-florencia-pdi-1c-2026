## ✦ Componentes Esenciales de Gradio

La API de Gradio cuenta con componentes prediseñados y configurados para interactuar de forma intuitiva con imágenes.

````python
import gradio as gr

# 1. COMPONENTES DE ENTRADA (Inputs)
gr.Image(type='pil')           # Entrada visual que pasa una imagen PIL al backend
gr.Textbox()                   # Campo de entrada de texto alfanumérico
gr.Number()                    # Caja de entrada numérica
gr.Slider(minimum=0, maximum=1) # Deslizador lineal analítico

# 2. COMPONENTES DE SALIDA (Outputs)
gr.Label()                     # Caja de clasificación con barras de confianza
gr.Image()                     # Renderizador visual para imágenes procesadas o recortadas
gr.Textbox()                   # Caja de salida textual para metadatos o coordenadas

# 3. ESTRUCTURAS DE DISEÑO (Layout)
gr.Interface()                 # Abstracción simplificada y rápida de UI
gr.Blocks()                    # Contenedor avanzado para flujos y diseños a medida
gr.Row()                       # Organización horizontal (fila)
gr.Column()                    # Organización vertical (columna)
````

## ✦ Patrón de 3 Capas Modular

Este es el patrón arquitectónico estructurado que aplicamos para garantizar la separación de responsabilidades y la legibilidad de nuestro software.

````python
import gradio as gr
from transformers import pipeline

# =====================================================================
# 1. DATA LAYER: Carga única del modelo en memoria
# =====================================================================
modelo_vit = pipeline('image-classification', model='google/vit-base-patch16-224')

# =====================================================================
# 2. BUSINESS LOGIC: Validación y lógica de control
# =====================================================================
def procesar_muestra(imagen):
    if imagen is None:
        return {'Error': 'Por favor, carguen una muestra.'}
        
    try:
        resultados = modelo_vit(imagen)
        salida_limpia = {}
        for res in resultados[:5]:
            salida_limpia[res['label']] = float(res['score'])
        return salida_limpia
    except Exception as e:
        return {'Error de Inferencia': str(e)}

# =====================================================================
# 3. PRESENTATION LAYER: Interfaz gráfica declarada con Blocks
# =====================================================================
with gr.Blocks(theme=gr.themes.Soft()) as demo_app:
    gr.Markdown('# Clasificador Didáctico ViT')
    
    with gr.Row():
        img_input = gr.Image(type='pil', label='Imagen de Entrada')
        lbl_output = gr.Label(label='Predicciones de Salida')
        
    btn_ejecutar = gr.Button('Procesar', variant='primary')
    btn_ejecutar.click(fn=procesar_muestra, inputs=img_input, outputs=lbl_output)

if __name__ == '__main__':
    demo_app.launch()
````
