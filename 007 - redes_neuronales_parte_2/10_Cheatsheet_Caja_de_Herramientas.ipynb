## 1. Recetario de Capas de Keras y Arquitecturas

A continuación, presentamos la sintaxis de las capas fundamentales para armar modelos de Perceptrón Multicapa (MLP) y Redes Neuronales Convolucionales (CNNs).

```python
import tensorflow as tf

# =========================================================================
# RECETA 1: ARQUITECTURA DE PERCEPTRÓN MULTICAPA (MLP)
# =========================================================================

modelo_mlp = tf.keras.Sequential([
    # Capa de Entrada: Especificamos la forma de la imagen (ej: 28x28 y 1 canal)
    tf.keras.layers.Input(shape=(28, 28, 1)),
    
    # Capa de Aplanado: Redimensiona de 2D (28x28) a vector plano (784)
    tf.keras.layers.Flatten(),
    
    # Capas ocultas totalmente conectadas (Dense) con activación ReLU
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    
    # Capa de salida: una neurona por clase y activación Softmax para probabilidades
    tf.keras.layers.Dense(units=10, activation='softmax')
])

# =========================================================================
# RECETA 2: ARQUITECTURA DE RED NEURONAL CONVOLUCIONAL (CNN)
# =========================================================================

modelo_cnn = tf.keras.Sequential([
    # Capa de Entrada: Imagen 2D sin aplanar
    tf.keras.layers.Input(shape=(28, 28, 1)),
    
    # Primera capa convolucional: 16 filtros de 3x3 y activación ReLU
    tf.keras.layers.Conv2D(filters=16, kernel_size=(3, 3), activation='relu'),
    # Primera capa de Pooling: reduce resolución a la mitad espacial
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    
    # Segunda capa convolucional: 32 filtros de 3x3 y activación ReLU
    tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    # Segunda capa de Pooling: reduce resolución espacial a la mitad
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    
    # Aplanamos los mapas de características jerárquicos reducidos
    tf.keras.layers.Flatten(),
    
    # Capa Dense intermedia para razonamiento abstracto
    tf.keras.layers.Dense(units=64, activation='relu'),
    
    # Capa de salida: probabilidades para las 10 categorías
    tf.keras.layers.Dense(units=10, activation='softmax')
])

print("✓ Plantillas de arquitecturas MLP y CNN instanciadas correctamente.")
```

## 2. Compilación y Configuración del Aprendizaje

La compilación define de qué manera va a aprender la red durante su optimización numérica.

```python
# =========================================================================
# CONFIGURACIÓN DE COMPILACIÓN ESTÁNDAR
# =========================================================================

modelo_cnn.compile(
    # Optimizador Adam: ajusta dinámicamente la tasa de aprendizaje
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    
    # Loss para problemas multiclase donde las etiquetas son números enteros (0, 1, 2...)
    loss='sparse_categorical_crossentropy',
    
    # Métrica para evaluar visualmente el rendimiento
    metrics=['accuracy']
)

print("✓ Configuración de aprendizaje y optimización compilada exitosamente.")
```

## 3. Receta para Gráficas de Curvas de Aprendizaje

Una función útil para dibujar rápidamente las curvas de pérdida y exactitud a partir del historial retornado por `model.fit()`.

```python
import matplotlib.pyplot as plt

def graficar_curvas_aprendizaje(historial_entrenamiento):
    """
    Dibuja de forma elegante y automatizada las métricas de precisión y pérdida.
    """
    historial_dict = historial_entrenamiento.history
    epocas = range(1, len(historial_dict['accuracy']) + 1)
    
    plt.figure(figsize=(12, 5))
    
    # Dibujamos la Precisión
    plt.subplot(1, 2, 1)
    plt.plot(epocas, historial_dict['accuracy'], 'bo-', label='Precisión')
    if 'val_accuracy' in historial_dict:
        plt.plot(epocas, historial_dict['val_accuracy'], 'ro-', label='Precisión Val')
    plt.title('Precisión durante el Aprendizaje', fontsize=12, fontweight="bold")
    plt.xlabel('Épocas')
    plt.ylabel('Precisión (Accuracy)')
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    
    # Dibujamos la Pérdida
    plt.subplot(1, 2, 2)
    plt.plot(epocas, historial_dict['loss'], 'gs-', label='Pérdida')
    if 'val_loss' in historial_dict:
        plt.plot(epocas, historial_dict['val_loss'], 'rs-', label='Pérdida Val')
    plt.title('Evolución de la Pérdida', fontsize=12, fontweight="bold")
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida (Loss)')
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    
    plt.tight_layout()
    plt.show()

print("✓ Función 'graficar_curvas_aprendizaje' compilada y disponible.")
```
